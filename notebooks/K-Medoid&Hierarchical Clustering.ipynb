# 2.1 K-Medoids & Hierarchical Clustering


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
import sys
sys.path.append("..")
from ml_model.kmedoids import SimpleKMedoids
import os

if os.path.exists('../data/raw_data.csv'):
    path = '../data/raw_data.csv'
else:
    path = 'data/raw_data.csv'

df = pd.read_csv(path).sample(2000, random_state=42)
if 'id' in df.columns:
    df.drop('id', axis=1, inplace=True)
if 'Response' in df.columns:
    df.drop('Response', axis=1, inplace=True)
```

## 2.2 Data Preprocessing for Clustering


```python
df.fillna(df.median(numeric_only=True), inplace=True)
for col in df.select_dtypes(include=['object']).columns:
    df[col].fillna(df[col].mode()[0], inplace=True)
    df[col] = LabelEncoder().fit_transform(df[col])

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

## 2.3 Applying K-Medoids


```python
kmedoids = SimpleKMedoids(n_clusters=3, max_iter=50, random_state=42)
kmedoids.fit(X_scaled)
kmed_labels = kmedoids.predict(X_scaled)
print("K-Medoids done.")
```

## 2.4 Applying Hierarchical Clustering


```python
hierarchical = AgglomerativeClustering(n_clusters=3, linkage='ward')
hier_labels = hierarchical.fit_predict(X_scaled)
print("Hierarchical Clustering done.")
```

## 2.5 PCA Projection


```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(X_pca[:,0], X_pca[:,1], c=kmed_labels, cmap='viridis')
plt.title("K-Medoids Clusters")

plt.subplot(1, 2, 2)
plt.scatter(X_pca[:,0], X_pca[:,1], c=hier_labels, cmap='plasma')
plt.title("Hierarchical Clusters")
plt.show()
```
